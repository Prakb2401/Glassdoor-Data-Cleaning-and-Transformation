```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string

# Download the set of stop words the first time
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Initialize a stemmer
stemmer = PorterStemmer()

def clean_description(description):
    # Lowercase the description
    description = description.lower()
    # Remove punctuation
    description = description.translate(str.maketrans('', '', string.punctuation))
    return description

# Apply the cleaning function to each job description
unclean_df_3['Job Description'] = unclean_df_3['Job Description'].apply(clean_description)
```

```python
unclean_df_3['Python']=unclean_df_3['Job Description'].apply(lambda x: 1 if "python" in x.lower() else 0)
unclean_df_3['Excel']=unclean_df_3['Job Description'].apply(lambda x: 1 if "excel" in x.lower() else 0)
unclean_df_3['SQL']=unclean_df_3['Job Description'].apply(lambda x: 1 if "sql" in x.lower() else 0)
unclean_df_3['Tableau']=unclean_df_3['Job Description'].apply(lambda x: 1 if "tableau" in x.lower() else 0)
unclean_df_3['Spark']=unclean_df_3['Job Description'].apply(lambda x: 1 if "spark" in x.lower() else 0)
unclean_df_3['Machine Learing']=unclean_df_3['Job Description'].apply(lambda x: 1 if "machine learning" in x.lower() else 0)
unclean_df_3['AWS']=unclean_df_3['Job Description'].apply(lambda x: 1 if "aws" in x.lower() else 0)
```
